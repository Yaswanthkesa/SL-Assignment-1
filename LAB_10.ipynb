```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score,f1_score
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df=pd.read_csv("spam.csv",encoding='latin1')
#latin1 encoding is used for assigning the unique numerical value to each character which includes many a non ascii characters as well
```

```python
df.head()
```

```python
df=df[['v1','v2']]
df.head()
```

```python
df=df.rename(columns={'v1':'label','v2':'text'})
```

```python
df.head()
```

```python
x=df['text']
y=df['label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#value_counts is used to count the number of unique values in a dataset
distrubution = y_train.value_counts()
distrubution
```

```python
distrubution.plot(kind='pie',autopct='%1.1f%%')
plt.title("Distribution of spam and ham messages")
plt.show()
```

```python
#generate word cloud for spam messages
spam_text=''.join(df[df['label']=='spam']['text'])
spam_text
```

```python
#generate the wordcloud for spam email
#a wordcloud in python is a  virtual representation of text data that uses the size and color of words to show their frequency
spam_wordcloud=WordCloud(width=800,height=400,max_words=100,background_color='white',random_state=42).generate(spam_text)
```

```python
spam_wordcloud
```

```python
#plot the word cloud for spam messages
plt.figure(figsize=(10,4))
plt.subplot(1,2,2)
plt.imshow(spam_wordcloud)
plt.title("Word Cloud for Spam Messages")
plt.axis('off')

#plot the word cloud for ham messages
plt.subplot(1,2,1)
plt.imshow(spam_wordcloud)
plt.title("Word Cloud for ham Messages")
plt.axis('off')
```

```python
#create a countvectorizer to covert text data
#numerical features
vectorizer=CountVectorizer()
x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
```

```python
#traina a multinomial naive bayes classifier
model_multinomial=MultinomialNB(alpha=0.8,fit_prior=True,force_alpha=True)
model_multinomial.fit(x_train,y_train)
```

```python
#train a guassian naive bayes classifier
model_gaussian=GaussianNB()
model_gaussian.fit(x_train.toarray(),y_train)
```

```python
#calculating the accuracy
y_pred_multinomial=model_multinomial.predict(x_test)
accuracy_multinomial=accuracy_score(y_test,y_pred_multinomial)
print("Accuracy of multinomial model is:" ,accuracy_multinomial)

y_pred_guassian=model_gaussian.predict(x_test.toarray())
accuracy_gaussian=accuracy_score(y_test,y_pred_guassian)
print("Accuracy of guassian naive bayes  model is:" ,accuracy_gaussian)

```

```python
methods=["Multinomial Naive Bayes","Guassian Naive Bayes"]
scores=[accuracy_multinomial,accuracy_gaussian]
plt.bar(methods,scores)
```
