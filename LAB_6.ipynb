```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("california-housing.csv")
```

```python
df.head()
```

```python
df.head(9)
```

```python
df["ocean_proximity"]=df.groupby("ocean_proximity")["households"].transform("mean")
```

```python
df
```

```python
#normalisation(scaling)
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df1=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
df1
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df1["households"]
x=df1.drop("households",axis=1)
y.shape,x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train.fillna(df.mean(),inplace=True)
x_test.fillna(df.mean(),inplace=True)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
linear_model=LinearRegression()
lasso_model=Lasso(alpha=0.1)
ridge_model=Ridge(alpha=0.1)
```

```python
lasso_model.fit(x_train,y_train)
```

```python
ridge_model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
