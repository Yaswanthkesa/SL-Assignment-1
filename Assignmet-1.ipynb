```python
def fibbo():
  fibonacci=[]
  n=int(input("Enter the limit: "))
  for i in range(0,n):
    if i==0:
        fibonacci.append(0)
    elif i==1:
        fibonacci.append(1)
    else:
        fibonacci.append(fibonacci[i-2]+fibonacci[i-1])
  print(fibonacci)
fibbo()
```

```python
def facto():
  n=int(input("Enter the number: "))
  fact=1
  for i in range(1,n+1):
      fact=fact*i
  print(fact)
facto()
```

```python
def sum_of_digits():
  n=int(input("Enter the number: "))
  sum=0
  while n>0:
    sum=sum+n%10
    n=n//10
  print(sum)
sum_of_digits()
```

```python
def prime():
  n=int(input("Enter the number: "))
  for i in range(2,n):
    if n%i==0:
      print("Not a prime number")
      break
  else:
    print("Prime number")
prime()
```

```python
def reverse_string():
  st=input("Enter the string: ")
  rev=""
  for i in range(len(st),0,-1):
    rev=rev+st[i-1]
  print(rev)
reverse_string()
```

```python
def check_palindrome():
  st=input("Enter the string : ")
  rev=""
  for i in range(len(st),0,-1) :
     rev=rev+st[i-1]
  if st==rev:
    print("Palindrome")
  else:
    print("Not a palindrome")
check_palindrome()
```

```python
def seperate():
   str=input("Enter the string:")
   str=str.lower()
   space=0
   vowels=['a','e','i','o','u']
   count=0
   for i in str:
     if i in ' ' :
       space = space+1
     if i in vowels:
       count=count+1
   print("Number of vowels in the string:",count)
   print("Number of consonants in the string:",len(str)-count-space)
seperate()
```

```python
def multiples():
   n=int(input("Enter the number:"))
   limit=int(input("Enter the limit:"))
   i=1
   mul = n*i
   while(mul<=limit):
     print(mul)
     i=i+1
     mul=n*i
multiples()
```
